```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = 2 + np.random.randn(100)
X2 = 2 + np.random.randn(100)
plt.scatter(X1, X2, color='r')

X1 = 10 + np.random.randn(100)
X2 = 10 + np.random.randn(100)
plt.scatter(X1, X2, color='b')

X_b = np.array([12, 0])
Y_b = np.array([0, 12])
plt.plot(X_b, Y_b, linestyle = '-.', color='green')
```

```python
n_points = 100
radius = 5
angle = np.random.rand(n_points) * 2 * np.pi
A1 = np.random.rand(n_points) * 0.7 * radius
x1 = np.cos(angle) * A1
y1 = np.sin(angle) * A1

A2 = np.random.rand(n_points) * radius + 1.5 * radius
x2 = np.cos(angle) * A2
y2 = np.sin(angle) * A2
```

```python
circle1 = plt.Circle((0, 0), radius=radius, color='black', fill=False)

fig, ax = plt.subplots()
fig = plt.gcf()
ax = fig.gca()
plt.xlim(-20, 20)
plt.ylim(-20, 20)

# ax.add_artist(circle1)
plt.scatter(x1, y1, color='blue')
plt.scatter(x2, y2, color='red')
```

```python
data1 = np.column_stack((x1, y1))
data1 = np.column_stack((data1, np.ones(data1.shape[0])[None].T))
data2 = np.column_stack((x2, y2))
data2 = np.column_stack((data2, np.zeros(data2.shape[0])[None].T))
data = np.concatenate((data1, data2))

data_0 = data[np.where(data[:, -1] == 0)]
data_1 = data[np.where(data[:, -1] == 1)]

X = data[:, :-1]
Y = data[:, -1]
```

```python
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.scatter(data_0[:, 0], data_0[:, 1], color='blue')
plt.scatter(data_1[:, 0], data_1[:, 1], color='red')
```
